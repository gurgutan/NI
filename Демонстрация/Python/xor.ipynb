```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import *
import matplotlib.pyplot as plt
```

```python
model = tf.keras.Sequential(
    [InputLayer(input_shape=(2,))]
    # + [Dense(8, activation=tf.nn.softsign) for i in range(1)]
    + [Dense(1), Activation(tf.nn.sigmoid),]
)

fit_rate = 0.1
fit_epochs = 256

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=fit_rate), loss="binary_crossentropy"
)
```

```python
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype="float32")
y = np.array([[0],    [1],    [1],    [0]], dtype="float32")
print(f"x=\n{x},\n\ny=\n{y}")
```

```python
history = model.fit(x, y, epochs=fit_epochs, verbose=1)
plt.plot(history.history["loss"])
plt.show()
```

```python
# Модель с 1 слоем Dense
model = tf.keras.Sequential(
    [InputLayer(input_shape=(2,))]
    + [Dense(8, activation=tf.nn.softsign) for i in range(1)]
    + [Dense(1), Activation(tf.nn.sigmoid),]
)
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=fit_rate), loss="binary_crossentropy"
)
history = model.fit(x, y, epochs=2048, verbose=1)
plt.plot(history.history["loss"])
plt.show()
```

```python
# Модель из 8 слоев Dense
fit_rate = 0.1
fit_epochs = 256
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype="float32")
y = np.array([[0],    [1],    [1],    [0]], dtype="float32")
print(f"x=\n{x},\n\ny=\n{y}")

model = tf.keras.Sequential(
    [InputLayer(input_shape=(2,))]
    + [Dense(8, activation=tf.nn.softsign) for i in range(64)]
    + [Dense(1), Activation(tf.nn.sigmoid),]
)
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=fit_rate), loss="binary_crossentropy"
)
history = model.fit(x, y, epochs=fit_epochs, verbose=1)
plt.plot(history.history["loss"])
plt.show()
```
