```python
import numpy as np
import math
```

```python
# Создадим случайные данные
x = np.linspace(-math.pi, math.pi, 2000)
y_true = np.sin(x)
print(f"y={y_true},\nx={x}")
print(y_true.shape, x.shape)
```

```python
# Случайные значения
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
d = np.random.randn()
a, b, c, d
```

```python
learning_rate = 1e-3
n = len(y_true)
for t in range(2**10):
    # Вычисление y по x
    # y = f(x) = z(x) = a + b x + c x^2 + d x^3
    y = a + b * x + c * x ** 2 + d * x ** 3

    # Вычисление средней ошибки 
    loss = np.square(y_true - y).sum() / n
    if t % 100 == 0:
        print(t, loss)

    # Вычисление градиента функции ошибки f'(z(x))*z'(x)
    grad_y_pred = -2.0 * (y_true - y) / n
    grad_a = grad_y_pred.sum() / n
    grad_b = (grad_y_pred * x).sum() / n
    grad_c = (grad_y_pred * x ** 2).sum() / n
    grad_d = (grad_y_pred * x ** 3).sum() / n

    # Update weights
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    c -= learning_rate * grad_c
    d -= learning_rate * grad_d

print(f'Result: y = {a} + {b} x + {c} x^2 + {d} x^3')
```
