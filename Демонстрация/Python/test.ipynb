```python
import torch.nn as nn
import torch
import matplotlib.pyplot as plt

def Radial(x: torch.Tensor, c: torch.Tensor, sigma: torch.float) -> torch.Tensor:
    alpha = nn.functional.sigmoid(sigma)*10+1  # [1, 11]
    y = torch.exp(-(x-c)**2/alpha)
    return y

c = torch.zeros((64,)).float()
x = torch.linspace(-2, 2, len(c))
sigma = 1
y = Radial(x, c, sigma)
y
```

```python
plt.plot(x, y)
plt.ylabel('radial')
plt.show()
```
