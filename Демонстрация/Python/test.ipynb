```python
import torch.nn as nn
import torch
import matplotlib.pyplot as plt

def Radial(x: torch.Tensor, c: torch.Tensor, sigma: torch.float) -> torch.Tensor:
    y = torch.exp(-(x-c)**2/sigma)
    return y

c = torch.zeros((100,)).float()
x = torch.linspace(-2, 2, len(c))
sigma = 1
y = Radial(x, c, sigma)

plt.plot(x, y)
plt.ylabel('radial')
plt.show()
```
